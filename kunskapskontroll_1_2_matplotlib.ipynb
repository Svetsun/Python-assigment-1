### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

The first approach is the functional method, which is quicker and simpler, but it offers less control over your visualizations. The other approach is the explicit object-oriented method.The explicit object-oriented method is recommended for use in Matplotlib. By doing so, you have more control over your visualizations. For instance, you have the ability to change parameters for visualizations.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

_**Figure**_ is the overall window or page where a plot is displayed. Multiple Axes objects can be created within a Figure by adjusting the argument in subplots(). 


_**Axes**_ is the area where data is visualized within an individual plot. It includes elements like titles, labels, and markers, and can hold multiple lines or data points. Using the explicit method of creating plots, you can specify the number of Axes you want.

When we talk about an _**Axis**_, we mean the x and y axis. They control the scale, limits, and markers along each axis.

An _**Artist**_  is any  object you can drow in Matplotlib , like Figures, Axes, Axis, or plot elements such as lines, text, and labels.

### When plotting in Matplotlib, what is the expected input data type?

In Matplotlib, we can use several data types as input, including lists, tuples, NumPy arrays, and Python dictionaries. You can also use Pandas Series and DataFrames when visualizing data while working with Pandas.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

Plot with object-oriented approach

```python
import numpy as np

x = np.linspace(-4, 4, 100)
y = x ** 2 
fig, ax = plt.subplots(figsize=( 6,6))
ax.plot(x, y)

ax.set_title("Function  y = x^2")
ax.set_xlabel("X")
ax.set_ylabel("Y")

```

Function (pyplot) oriented approach

```python
x = np.linspace(-4, 4, 100)
y = x ** 2  
plt.plot(x, y)

plt.title("Function y = x^2")
plt.xlabel("X")
plt.ylabel("Y")
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig,axs = plt.subplots(1, 2, figsize=(12, 5))

# Scatter plot
axs[0].scatter(random_data_x, random_data_y, alpha=0.5)
axs[0].set_title("Scatter Plot")
axs[0].set_xlabel("X")
axs[0].set_ylabel("Y")

# Bar plot
axs[1].bar(names, values, color='skyblue')
axs[1].set_title("Bar Plot")
axs[1].set_xlabel("Fruit")
axs[1].set_ylabel("Count")
```
