### Explain what a CSV file is.

#### CSV ( Comma-Separated Values fil)
**_CSV_** is the plain text file that stores data in a tabular format
+ **_Each line represents a row in the table_**
+ **_Each column is separated by a comma (or sometimes another delimiter, like a semicolon)_**
+ **_The first row often contains column headers that describe the data in each column_**
